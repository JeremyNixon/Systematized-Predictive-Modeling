```python
class classifier():
    def __init__(self, x, y, hyperparameters='individual'):
        self.hyperparameters = hyperparameters
        if type(y[0]) != int:
            y, self.mapping = self.categorize(y)
        self.x_train, self.x_test, self.y_train, self.y_test = sklearn.cross_validation.train_test_split\
            (np.array(x), np.array(y, dtype = int), test_size = .20, random_state=42)
        self.x_train1 = self.x_train[:len(self.x_train)/2]
        self.x_train2 = self.x_train[len(self.x_train)/2:]
        self.y_train1 = self.y_train[:len(self.x_train)/2]
        self.y_train2 = self.y_train[len(self.x_train)/2:]
    
    def accuracy(self, predictions):
        count = 0
        for i in xrange(len(predictions)):
            if predictions[i] == self.y_test[i]:
                count += 1
        return count/float(len(predictions))
    
    def categorize(self, labels):
        mapping = {}
        labels = pd.Series(labels)
        for i,v in enumerate(list(set(labels))):
            labels = labels.replace(v, i)
            mapping[i] = v
        return labels, mapping  
    
    def stacker(self):
        stacked_data = np.append(self.x_train2, gradient_booster(self.x_train1, self.y_train1, self.x_train2), 1)
        return logistic_regression(stacked_data, y_train2, x_test)
    
    def xgboost(self):
        dtrain = xgb.DMatrix(self.x_train, label=self.y_train)
        dtest = xgb.DMatrix(self.x_test)
        param = {'bst:max_depth':50, 'bst:eta':1, 'silent':1,'objective':'multi:softmax'}
        param['num_class'] = self.x_train.shape[1]
#         evallist = [(dtest,'eval'), (dtrain,'train')]
        num_round = 10
        bst = xgb.train(param, dtrain, num_round, verbose_eval=False)
        preds = bst.predict(dtest)
#         for i in xrange(len(labels)):
#             print labels[i], self.y_test[i]
        return preds
    
    def gradient_booster(self, x_train = self.x_train, y_train = self.y_train,\
                         x_test = self.x_test, y_test = self.y_test, n_estimators=100):
        if self.hyperparameters == 'default':
            booster = ensemble.GradientBoostingClassifier(n_estimators = 200)
            booster = booster.fit(x_train, y_train)
            booster_predictions = booster.predict(x_test)
            return booster_predictions
        elif self.hyperparameters == 'individual':
            max_accuracy = -1
            for n_estimators in xrange(100, 1300, 300):
                booster = ensemble.GradientBoostingClassifier\
                (n_estimators = n_estimators)
                booster = booster.fit(x_train, y_train)
                booster_predictions = booster.predict(x_test)
                if self.accuracy(booster_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(booster_predictions)
                    optimal_parameter = n_estimators
            optimal_n_estimators = optimal_parameter
            max_accuracy = -1
            for max_depth in xrange(2,15,3):
                booster = ensemble.GradientBoostingClassifier\
                (max_depth = max_depth)
                booster = booster.fit(x_train, y_train)
                booster_predictions = booster.predict(x_test)
                if self.accuracy(booster_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(booster_predictions)
                    optimal_parameter = max_depth
            optimal_max_depth = optimal_parameter
            print "Optimal Booster Hyperparams: n_estimators = %d, max_depth = %d" \
                    %(optimal_n_estimators, optimal_max_depth)
            booster = ensemble.GradientBoostingClassifier(\
                          n_estimators = optimal_n_estimators,
                          max_depth = optimal_max_depth)
            booster = booster.fit(x_train, y_train)
            booster_predictions = booster.predict(x_test)
            return booster_predictions
        elif self.hyperparameters == 'grid':
            max_accuracy = -1
            for n_estimators in xrange(100, 1000, 300):
                for max_depth in xrange(2,15,3):
                    booster = ensemble.GradientBoostingClassifier\
                    (n_estimators = n_estimators, max_depth = max_depth)
                    booster = booster.fit(x_train, y_train)
                    booster_predictions = booster.predict(x_test)
                    print str(self.accuracy(booster_predictions)) + ' ' + str(n_estimators) + ' ' + str(max_depth)
                    if self.accuracy(booster_predictions) > max_accuracy:
                        max_accuracy = self.accuracy(booster_predictions)
                        max_predictions = booster_predictions
                        optimal_parameters = [n_estimators, max_depth]
            print("Gradient booster: ", optimal_parameters)
            return max_predictions
                
    
    def random_forest(self, n_estimators=100):
        if self.hyperparameters == 'default':
            forest = ensemble.RandomForestClassifier(n_estimators)
            forest = forest.fit(self.x_train, self.y_train)
            forest_predictions = forest.predict(self.x_test)
            return forest_predictions
        elif self.hyperparameters == 'individual':
            max_accuracy = -1
            for n_estimators in xrange(100, 1400, 400):
                forest = ensemble.RandomForestClassifier(n_estimators = n_estimators)
                forest = forest.fit(self.x_train, self.y_train)
                forest_predictions = forest.predict(self.x_test)
                if self.accuracy(forest_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(forest_predictions)
                    optimal_parameter = n_estimators
            optimal_n_estimators = optimal_parameter
            feat = self.x_train.shape[1]
            max_accuracy = -1
            for proportion in xrange(int(.3*feat), int(.6*feat), int(.05*feat)+1):
                forest = ensemble.RandomForestClassifier(max_features = proportion)
                forest = forest.fit(self.x_train, self.y_train)
                forest_predictions = forest.predict(self.x_test)
                if self.accuracy(forest_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(forest_predictions)
                    optimal_parameter = proportion
            optimal_max_features = optimal_parameter
            print "Optimal Forest Hyperparams: n_estimators = %d, max_features = %d" \
                    %(optimal_n_estimators, optimal_max_features)
            forest = ensemble.RandomForestClassifier(\
                 n_estimators=optimal_n_estimators, max_features = optimal_max_features)
            forest = forest.fit(self.x_train, self.y_train)
            forest_predictions = forest.predict(self.x_test)
            return forest_predictions 
        elif self.hyperparameters == 'grid':
            feat = self.x_train.shape[1]
            max_accuracy = -1
            for n_estimators in xrange(100, 1000, 200):
                for proportion in xrange(int(.3*feat), int(.6*feat), int(.03*feat)+1):
                    for criterion in ['gini', 'entropy']:
                        forest = ensemble.RandomForestClassifier(n_estimators = n_estimators, criterion = criterion, max_features = proportion)
                        forest = forest.fit(self.x_train, self.y_train)
                        forest_predictions = forest.predict(self.x_test)
                        print self.accuracy(forest_predictions)
                        if self.accuracy(forest_predictions) > max_accuracy:
                            max_accuracy = self.accuracy(forest_predictions)
                            max_predictions = forest_predictions
                            optimal_parameters = [n_estimators, proportion, criterion]
            print["Random forest: ", optimal_parameters]
            return max_predictions
                    
    
    def extremely_random_forest(self, n_estimators=100):
        if self.hyperparameters == 'default':
            erf = ensemble.ExtraTreesClassifier(n_estimators)
            erf = erf.fit(self.x_train, self.y_train)
            erf_predictions = erf.predict(self.x_test)
            return erf_predictions
        elif self.hyperparameters == 'individual':
            max_accuracy = -1
            for n_estimators in xrange(100, 1300, 400):
                erf = ensemble.ExtraTreesClassifier(n_estimators = n_estimators)
                erf = erf.fit(self.x_train, self.y_train)
                erf_predictions = erf.predict(self.x_test)
                if self.accuracy(erf_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(erf_predictions)
                    optimal_parameter = n_estimators
            optimal_n_estimators = optimal_parameter
            feat = self.x_train.shape[1]
            max_accuracy = -1
            for proportion in xrange(int(.3*feat), int(.6*feat), int(.05*feat)+1):
                erf = ensemble.ExtraTreesClassifier(max_features = proportion)
                erf = erf.fit(self.x_train, self.y_train)
                erf_predictions = erf.predict(self.x_test)
                if self.accuracy(erf_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(erf_predictions)
                    optimal_parameter = proportion
            optimal_proportion = optimal_parameter
            print "Optimal erf Hyperparams: n_estimators = %d, proportion = %d" \
                    %(optimal_n_estimators, optimal_proportion)
            erf = ensemble.ExtraTreesClassifier(\
                 n_estimators=optimal_n_estimators, max_features = optimal_proportion)
            erf = erf.fit(self.x_train, self.y_train)
            erf_predictions = erf.predict(self.x_test)
            return erf_predictions
        
    def logistic_regression(self, x_train = self.x_train, y_train = self.y_train,\
                         x_test = self.x_test, y_test = self.y_test):
        if self.hyperparameters == 'default':
            logistic = linear_model.LogisticRegression(penalty="l2")#class_weight='auto'
            logistic.fit(x_train, y_train)
            logistic_predictions = logistic.predict(x_test)
            return logistic_predictions
        elif self.hyperparameters == 'individual':
            for penalty in ['l1', 'l2']:
                if penalty == 'l1':
                    logistic = linear_model.LogisticRegression(penalty="l1")#class_weight='auto'
                    logistic.fit(x_train, y_train)
                    logistic_predictions = logistic.predict(x_test)
                    l1_accuracy = self.accuracy(logistic_predictions)
                    optimal_penalty = 'l1'
                else:
                    logistic = linear_model.LogisticRegression(penalty="l2")#class_weight='auto'
                    logistic.fit(x_train, y_train)
                    logistic_predictions = logistic.predict(x_test)
                    if self.accuracy(logistic_predictions) > l1_accuracy:
                        optimal_penalty = 'l2'
            max_accuracy = -1
            for C in xrange(1,104,4):
                logistic = logistic = linear_model.LogisticRegression(penalty=optimal_penalty, C=C)
                logistic.fit(x_train, y_train)
                logistic_predictions = logistic.predict(x_test)
                if self.accuracy(logistic_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(logistic_predictions)
                    optimal_parameter = C
            optimal_C = optimal_parameter
            print "Optimal Logistic Regression Hyperparams: Penalty = %s, C = %d."\
                %(optimal_penalty, optimal_C)
            logistic = linear_model.LogisticRegression(penalty=optimal_penalty, C = optimal_C)#class_weight='auto'
            logistic.fit(x_train, y_train)
            logistic_predictions = logistic.predict(x_test)
            return logistic_predictions
    
    def lda(self):
        lda = discriminant_analysis.LinearDiscriminantAnalysis()
        lda.fit(self.x_train, self.y_train)
        lda_predictions = lda.predict(self.x_test)
        return lda_predictions
    
    def qda(self):
        if self.hyperparameters == 'default':
            qda = discriminant_analysis.QuadraticDiscriminantAnalysis()
            qda.fit(self.x_train, self.y_train)
            qda_predictions = qda.predict(self.x_test)
            return qda_predictions
        elif self.hyperparameters == 'individual':
            max_accuracy = -1
            for reg_param in [x*.1 for x in range(0,10,2)]:
                qda = discriminant_analysis.QuadraticDiscriminantAnalysis(reg_param = reg_param)
                qda.fit(self.x_train, self.y_train)
                qda_predictions = qda.predict(self.x_test)
                if self.accuracy(qda_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(qda_predictions)
                    optimal_parameter = reg_param
            optimal_reg_param = optimal_parameter
            print "Optimal QDA Hyperparams: Reg_param = %d" %(optimal_reg_param)
            qda = discriminant_analysis.QuadraticDiscriminantAnalysis(reg_param = optimal_reg_param)
            qda.fit(self.x_train, self.y_train)
            qda_predictions = qda.predict(self.x_test)
            return qda_predictions
    
    def svc(self):
        if self.hyperparameters == 'default':
            support_vm = svm.SVC()
            support_vm.fit(self.x_train, self.y_train)
            svm_predictions = support_vm.predict(self.x_test)
            return svm_predictions
        elif self.hyperparameters == 'individual':
            max_accuracy = -1
            for C in [10, 100]:
                s = svm.SVC(C=C)
                s.fit(self.x_train, self.y_train)
                svm_predictions = s.predict(self.x_test)
                if self.accuracy(svm_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(svm_predictions)
                    optimal_parameter = C
            optimal_C = optimal_parameter
            max_accuracy = -1
            for kernel in ['rbf', 'linear', 'poly']:
                s = svm.SVC(C = optimal_C, kernel=kernel)
                s.fit(self.x_train, self.y_train)
                svm_predictions = s.predict(self.x_test)
                if self.accuracy(svm_predictions) > max_accuracy:
                    max_accuracy = self.accuracy(svm_predictions)
                    optimal_parameter = kernel
            optimal_kernel = optimal_parameter
            print "Optimal SVM Hyperparams: Kernel = %s, C = %d" %(optimal_kernel, optimal_C)
            s = svm.SVC(C = optimal_C, kernel=optimal_kernel)
            s.fit(self.x_train, self.y_train)
            svm_predictions = s.predict(self.x_test)
            return svm_predictions
    
    def blend(self, *args):
        n = len(args)
        blended_predictions = []
        for i in xrange(len(args[0])):
            counts = {}
            for j in xrange(n):
                try:
                    counts[args[j][i]] += 1
                except KeyError:
                    counts[args[j][i]] = 1
            maximus = -float('inf')
            for index, value in counts.iteritems():
                if value > maximus:
                    max_value = index
                    maximus = value
            blended_predictions.append(max_value)
        return blended_predictions

    def predict(self):
        xgboost_predictions = self.xgboost()
        booster_predictions = self.gradient_booster()
        forest_predictions = self.random_forest()
        erf_predictions = self.extremely_random_forest()
        logistic_predictions = self.logistic_regression()
        lda_predictions = self.lda()
        qda_predictions = self.qda()
#         svm_predictions = self.svc()
        blended_predictions = self.blend(xgboost_predictions, booster_predictions, forest_predictions, erf_predictions,\
                  logistic_predictions, lda_predictions, qda_predictions)
        
        models = ['xgboost_predictions','booster_predictions', 'forest_predictions', 'erf_predictions',\
                  'logistic_predictions', 'lda_predictions', 'qda_predictions', \
                  'blended_predictions']
        for i, v in enumerate([xgboost_predictions, booster_predictions, forest_predictions, erf_predictions,\
                  logistic_predictions, lda_predictions, qda_predictions,\
                 blended_predictions]):
        
            print self.accuracy(v), models[i]
        
        return [xgboost_predictions, booster_predictions, forest_predictions, erf_predictions,\
                  logistic_predictions, lda_predictions, qda_predictions,\
                 blended_predictions, self.y_test]
```
